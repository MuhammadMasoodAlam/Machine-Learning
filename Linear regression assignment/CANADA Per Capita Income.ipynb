```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from pandas.core.common import random_state
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df_ci=pd.read_csv("canada_per_capita_income.csv")
```

```python
plt.title("Per Capita Income")
sns.distplot(df_ci["per capita income (US$)"])
plt.show()
```

```python
# Relationship between Years & Inome
plt.scatter(df_ci["year"],df_ci["per capita income (US$)"],color="Red")
plt.title("Per Capita Income Over The Years")
plt.xlabel('Years')
plt.ylabel("Per Capilta Income")
plt.show()
```

```python
X=df_ci.iloc[:,:1]
y=df_ci.iloc[:,1:]
```

```python
# Spilitting Data into Features & target Variable
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.20,random_state=40)
```

```python
X_train
```

```python
y_train
```

```python
X_test
```

```python
y_test
```

```python
reg=LinearRegression()
reg.fit(X_train,y_train)
```

```python
# Predict Results
y_pred_test=reg.predict(X_test) # predict the value of y_test
y_pred_train=reg.predict(X_train) # predict the value of y_train
```

```python
y_pred_test
```

```python
y_pred_train
```

```python
# Visualize Prediction
```

```python
plt.scatter(df_ci['year'],df_ci["per capita income (US$)"], color="purple")
plt.plot(X_train,y_pred_train,color="Blue")
plt.title("Per Capita Income Over The Years (Test Data)")
plt.xlabel("Years")
plt.ylabel("Per Capita Income")
plt.legend(["X-test/y_test", "X_train/Y_pred_train"],title="Income/Years",loc="best",facecolor="Red")
plt.show()
```

```python
# Coefficient & Intercept
print(f'Coefficient: {reg.coef_}')
print(f'Intercept:  {reg.intercept_}')
print("y= ", reg.intercept_,"+",reg.coef_,"* X")
```

```python
# Evaluate the Model
print("MSME = ", mean_squared_error(y_test,y_pred_test))
print("R2 = ", r2_score(y_test,y_pred_test))
print("RMSE = ", np.sqrt(mean_squared_error(y_test,y_pred_test)))
```

```python
print("accuracy = ", int(r2_score(y_test,y_pred_test)*100))
```
